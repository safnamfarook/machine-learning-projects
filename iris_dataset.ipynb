# **Building a Classification Model for Iris Dataset**

# Import Libraries

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
```

# Load Dataset


```python
iris = datasets.load_iris()
iris
```

## Input Features

```python
iris.data
```

```python
print (iris.feature_names)
```

## Output Features

```python
iris.target
```

```python
print (iris.target_names)
```

# Glimpse of the data

## Assigning input and output variables

```python
X = iris.data
y = iris.target
```

## Examining data dimension

```python
print (X.shape)
print (y.shape)
```

# Build Classification Model using Random Forest

```python
clf = RandomForestClassifier()
```

```python
clf.fit(X,y)
```

# Feature Importance

```python
print (clf.feature_importances_)
```

# Make Prediction

```python
X[0]
```

```python
print (clf.predict([[5.1, 3.5, 1.4, 0.2]]))
```

```python
print (clf.predict_proba(X[[0]]))
```

```python
clf.fit(iris.data, iris.target_names[iris.target])
```

# Data Split (80/20 ratio)

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

# Rebuild the Random Forest Model

```python
clf.fit(X_train, y_train)
```

## Performs prediction on single sample from the data set

```python
print (clf.predict([[5.1, 3.5, 1.4, 0.2]]))
```

```python
print (clf.predict_proba([[5.1, 3.5, 1.4, 0.2]]))
```

## Performs prediction on the test set

### Predicted class labels

```python
print (clf.predict(X_test))
```

### Actual class labels

```python
print (y_test)
```

# Model Performance

```python
print (clf.score(X_test, y_test))
```
